## Insights

```python
# Há diferença entre a quantidade de bicicletas alugadas nos dias úteis e não úteis?

plt.bar(df_day['workingday'], df_day['cnt'])
plt.xticks([0, 1], ['Dia Útil', 'Dia Não Útil'])
plt.xlabel('Dias úteis e não úteis')
plt.ylabel('Quantidade de Bicicletas Alugadas')
plt.title('Quantidade de bicicletas alugadas nos dias úteis e não úteis')
plt.show()
```

```python
# Existe correlação da condição climática com a quantidade de bicicletas alugadas?


# Verificar a distribuição da quantidade de bicicletas alugadas em relação à condição climática
plt.bar(df_day['weathersit'], df_day['cnt'])
plt.xticks ([1, 2, 3], ['Céu limpo', 'Nublado', 'Chuva fraca'])
plt.xlabel('Condição Climática')
plt.ylabel('Quantidade de Bicicletas Alugadas')
plt.title('Relação entre Condição Climática e Aluguéis de Bicicletas')
plt.show()

# Calcular a correlação entre condição climática e quantidade de bicicletas alugadas
correlacao = df_day['weathersit'].corr(df_day['cnt'])

print(f"Correlação entre condição climática e quantidade de bicicletas alugadas: {correlacao}")
```

```python
# Renomear estações
estacoes = {
    1: 'Inverno',
    2: 'Primavera',
    3: 'Verão',
    4: 'Outono'
}

df_day['season'] = df_day['season'].map(estacoes)

# Calcular a média de aluguéis para cada estação
media_por_estacao = df_day.groupby('season')['cnt'].mean()

# Plotar um gráfico de barras para visualizar a média de aluguéis por estação
plt.figure(figsize=(8, 5))
media_por_estacao.plot(kind='bar')
plt.xlabel('Estação do Ano')
plt.ylabel('Média de Aluguéis')
plt.title('Média de Aluguéis de Bicicletas por Estação do Ano')
plt.xticks(rotation=45)
plt.show()
```
