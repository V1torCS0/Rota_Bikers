## Perguntas de partida e hipóteses

A partir das informações levantadas, podemos supor que:

- Há diferença entre a quantidade de bicicletas alugadas nos dias úteis e não úteis?
- Existe correlação da condição climática com a quantidade de bicicletas alugadas?
- As estações do ano podem influenciar na quantidade de bicicletas alugadas?
- A quantidade de usuários casuais e registrados oscila ao longo do tempo?

Analisando as variáveis das perguntas acima, somos capazes de entender se existe uma correlação entre os fatores externos e a nossa variável alvo `Cnt`.

## Insights

```python
# Há diferença entre a quantidade de bicicletas alugadas nos dias úteis e não úteis?
resposta = {
    1: 'Dia útil',
    0: 'Dia não útil'}

df_day['workingday'] = df_day['workingday'].map(resposta)

# Media da quantidade de aluguéis por resposta
media = df_day.groupby('workingday')['cnt'].mean()

media.plot(kind='barh')
plt.title('Quantidade de bicicletas alugadas nos dias úteis e não úteis')
plt.ylabel('')
plt.show()
```

```python
# Existe correlação da condição climática com a quantidade de bicicletas alugadas?

condicoes_climaticas = {
    1: 'Céu limpo',
    2: 'Nublado',
    3: 'Chuva fraca',
    4: 'Chuva forte'}

df_day['weathersit'] = df_day['weathersit'].map(condicoes_climaticas)

# Calcular a quantidade de aluguéis para cada estação
condicoes_climaticas = df_day.groupby('weathersit')['cnt'].mean()

# Verificar a distribuição da quantidade de bicicletas alugadas em relação à condição climática
ordem = ['Céu limpo', 'Nublado', 'Chuva fraca', 'Chuva forte']
condicoes_climaticas = condicoes_climaticas.reindex(ordem)

condicoes_climaticas.plot(kind='barh')
plt.title('Quantidade de bicicletas alugadas')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.show()
```

```python
# A estação do ano influencia na quantidade de bicicletas alugadas?

# Renomear estações
estacoes = {
    1: 'Inverno',
    2: 'Primavera',
    3: 'Verão',
    4: 'Outono'}

df_day['season'] = df_day['season'].map(estacoes)

# Calcular a média de aluguéis para cada estação
media_por_estacao = df_day.groupby('season')['cnt'].mean()

# Plotar um gráfico de barras para visualizar a média de aluguéis por estação
ordem = ['Verão', 'Outono', 'Inverno', 'Primavera']
media_por_estacao = media_por_estacao.reindex(ordem)

plt.figure(figsize=(8, 5))
media_por_estacao.plot(kind='barh')
plt.title('Média de Aluguéis de Bicicletas por Estação do Ano')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.show()
```
