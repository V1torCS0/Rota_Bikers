## Insights

```python
# Há diferença entre a quantidade de bicicletas alugadas nos dias úteis e não úteis?
resposta = {
    1: 'Dia útil',
    0: 'Dia não útil'}

df_day['workingday'] = df_day['workingday'].map(resposta)

# Media da quantidade de aluguéis por resposta
media = df_day.groupby('workingday')['cnt'].mean()

media.plot(kind='barh')
plt.title('Média da quantidade de bicicletas alugadas nos dias úteis e não úteis')
plt.ylabel('')
plt.show()
```

**Não há uma diferença significativa na média de aluguéis entre os dias úteis e não úteis.**

```python
# Existe correlação da condição climática com a quantidade de bicicletas alugadas?

condicoes_climaticas = {
    1: 'Céu limpo',
    2: 'Nublado',
    3: 'Chuva fraca',
    4: 'Chuva forte'}

df_day['weathersit'] = df_day['weathersit'].map(condicoes_climaticas)

# Calcular a quantidade de aluguéis para cada estação
condicoes_climaticas = df_day.groupby('weathersit')['cnt'].mean()

# Verificar a distribuição da quantidade de bicicletas alugadas em relação à condição climática
ordem = ['Céu limpo', 'Nublado', 'Chuva fraca', 'Chuva forte ou nevasca']
condicoes_climaticas = condicoes_climaticas.reindex(ordem)

condicoes_climaticas.plot(kind='barh')
plt.title('Quantidade de bicicletas alugadas x condição climática')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.show()
```

**As condições climáticas costumam impactar fortemente na quantidade de aluguéis por dia. Nos dias de céu limpo, a média de aluguéis é de quase 5000, enquanto não há qualquer registro de aluguel nos dias de chuva forte ou nevasca**

```python
# A estação do ano influencia na quantidade de bicicletas alugadas?

# Renomear estações
estacoes = {
    1: 'Inverno',
    2: 'Primavera',
    3: 'Verão',
    4: 'Outono'}

df_day['season'] = df_day['season'].map(estacoes)

# Calcular a média de aluguéis para cada estação
media_por_estacao = df_day.groupby('season')['cnt'].mean()

# Plotar um gráfico de barras para visualizar a média de aluguéis por estação
ordem = ['Verão', 'Outono', 'Inverno', 'Primavera']
media_por_estacao = media_por_estacao.reindex(ordem)

plt.figure(figsize=(8, 5))
media_por_estacao.plot(kind='barh')
plt.title('Média de Aluguéis de Bicicletas por Estação do Ano')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.show()
```

**Percebe-se que durante o inverno, quando as temperaturas do país despencam, a média dos aluguéis de bicicletas também diminui. Isso pode ser explicado por conta das nevascas que ocorrem em algumas regiões do país, situação que pode ser observada no gráfico 'Quantidade de bicicletas alugadas x condição climática', em que não há ocorrencia de aluguéis nesses dias.**

```python
# A quantidade de bicicletas alugadas aumentou ao longo do tempo?

dados_agrupados = df_day.groupby(['yr', 'mnth'])['cnt'].sum().unstack()

# Gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(dados_agrupados.loc[0], label='2011', marker='o')
plt.plot(dados_agrupados.loc[1], label='2012', marker='o')

plt.title('Comparativo de aluguéis de bicicleta durante os anos')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Aluguéis')
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True)
plt.show()
```

**As estações do ano nos EUA se dividem da seguinte forma:**

Inverno: de 12 a 2

Primavera: de 3 a 5

Verão: de 6 a 8

Outono: de 9 a 11

Então, é possível concluir que durante os meses mais frios, há uma grande queda na quantidade de bicicletas alugadas, a alta começa a ocorrer durante a primavera e atinge sua alta entre o verão e o outono.

```python
# A quantidade de usuários casuais e registrados oscila ao longo do tempo?

dados_agrupados_casual = df_day.groupby(['yr', 'mnth'])['casual'].sum().unstack()
dados_agrupados_registered = df_day.groupby(['yr', 'mnth'])['registered'].sum().unstack()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5), sharey=True)

# Subplot ano 0
ax1.plot(dados_agrupados_casual.loc[0], label='casual', marker='o', linestyle='-')
ax1.plot(dados_agrupados_registered.loc[0], label='Registered', marker='o', linestyle='-')
ax1.set_title('2011')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Quantidade de Aluguéis')
ax1.set_xticks(range(1, 13))
ax1.set_ylim(0, 180000)  # Define os limites do eixo Y
ax1.grid(True)
ax1.legend()

# Subplot ano 1
ax2.plot(dados_agrupados_casual.loc[1], label='casual', marker='o', linestyle='-')
ax2.plot(dados_agrupados_registered.loc[1], label='Registered', marker='o', linestyle='-')
ax2.set_title('2012')
ax2.set_xlabel('Mês')
ax2.set_ylabel('Quantidade de Aluguéis')
ax2.set_xticks(range(1, 13))
ax2.set_ylim(0, 180000)  # Define os limites do eixo Y
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

**A quantidade de usuários registrados aumentou significativamente durante o segundo ano, isso pode ter ocorrido por conta da popularização do serviço. Pelo gráfico também é possível perceber que a média de aluguéis mensais passou por um pequeno aumento para os usuários não registrados.**
